### Librerias adicionales

Para a√±adirle funciones adicionales a la placa tendremos que instalar algunas librerias adicionales.

Las librerias estan en:   
https://circuitpython.org/libraries

Hay dos paquetes zip el de adafruit y el de la comunidad.

1) Nos descargamos el adafruit-circuitpython-bundle-9.x   
   
   IMPORTANTE: la Version de las librerias tienen que ser la misma que la version de circuitpython instalada.   



2) Instalamos las librerias en el dispositivo.   
   Nota: Hay varias formas de instalar las librerias en el dispositivo.   
   La mas intuitiva es habilitar el modo "disco USB" / "USB Storage"    
   y arrastrar las librerias dentro de la carpeta lib.    
   (Tambien de puede hacer desde thonny o desde la interfaz web (apartado 3)
   
   En nuestro caso arrastramos la carpeta adafruit_hid.   
   Y el archivo adafruit_request.mpy

3) con esto ya podremos programar funciones adicionales    
para que el dispositivo haga de HID (teclado / raton)    
Y para que el dispositivo pueda comunicarse con Internet via API Requests
