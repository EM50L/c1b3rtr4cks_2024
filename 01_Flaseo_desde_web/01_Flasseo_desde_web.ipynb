Flassear/Burn un dispositivo electronico es proceso de escribir un firmware en la memoria interna(flash/Prom) de un dispositivo electrónico.

- Practicamente Cada dispositivo se programa de una forma distinta.
  Aunque cada vez se esta haciendo mas sencillo gracias a la evolucion de los bootloaders.

- En el contrexto de los microcontroladores:

  -- El "bootloader" es un programa que se ejecuta en el proceso de arranque del micro.    
  Cullo objetivo para cargar el núcleo del firmware y permitir su actualizacion.

  -- Bottloaders conocidos: Android , Arduino ...

- En las ultimas Generacion de Bootloaders.   
  Si arrancamos el dispositivo mientras mantenemos el boton boot pulsado.   
  El dispositivo electronico se presenta como un disco/pendrive.    
  Y para flassear/cargar el firmware solo hay que copiar el archivo en dicho disco.   
 

- Los ESP no suelen tener este tipo de bootloaders de serie.   
  Luego tenermos que recurrir a una herramienta llamada esptool
